```python
cap = cv2.VideoCapture("Red_Mountain_South.mp4")

# cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))



if cap.isOpened() == False:
    print("Error file not found")


while cap.isOpened():
    ret, frame = cap.read()
    
    if ret == True:
        # time.sleep(1/20)
        
        
        ## api call
        
        # frame to base64
        cv2.imwrite(filename="temp.jpg",img=frame)
        image64 = image_2_base64("temp.jpg").decode("utf-8")
        #print(image64)
        res = requests.post(url="http://127.0.0.1:5000/predictions_via_api",json={"image":image64})
        base64_2_img(res.content,"temp.jpg")
        img = cv2.imread("temp.jpg",1)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        
        
        
        img = cv2.resize(frame,(1024,640))
        
        
        
        
        cv2.imshow('Camera feed', img)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
        
```
